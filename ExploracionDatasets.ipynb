```python
import re

## Tratamos la columna rango_edad

def haz_rangos(frase):
    if frase not in ["Desconocido","<5",">74"]:
        return re.findall(r'\d+',frase)[0]+"-"+re.findall(r'\d+',frase)[1]
    else:
        return frase

# df.rango_edad[df.rango_edad=="Desconocido"] = "-1"
df.rango_edad[df.rango_edad=="Menor de 5 años"] = "<5"
df.rango_edad[df.rango_edad=="Más de 74 años"] = ">74"
df["rango_edad"] = df.rango_edad.transform(haz_rangos)

df.rango_edad.unique()
```

```python
## borramos las columnas que aportan informacion que no nos interesa

df = df.drop(['coordenada_x_utm', 'coordenada_y_utm'], axis=1)
df.columns
```

```python
## Procesamos las columnas "fecha" y "hora" a "fecha_hora"

df["fecha_hora"] = df.fecha.str.cat(df.hora, sep=" ")
```

```python
from datetime import datetime
def fecha_hora_to_seconds(fecha_hora_string):
    return datetime.strptime(fecha_hora_string, '%d/%m/%Y %H:%M:%S').strftime("%s")
df["fecha_hora"] = df.fecha_hora.transform(fecha_hora_to_seconds)
```

```python
import matplotlib.pyplot as plt
color_palette = ['skyblue', 'lightgreen', 'gold']
## 3.Hacer análisis sobre las fechas en las que más accidentes se producen
accidentes = df.drop_duplicates(["num_expediente"], keep='first').fecha.value_counts().to_dict()
plt.barh(list(accidentes.keys())[0:10], list(accidentes.values())[0:10], color=color_palette)
plt.title("Top 10 dias con más accidentes de tráfico")
plt.xlabel("Numero accidentes")
plt.ylabel("Fechas")
plt.show()
```

```python
plt.barh(list(accidentes.keys())[-10:], list(accidentes.values())[-10:], color=color_palette)
plt.title("Top 10 dias con menos accidentes de tráfico")
plt.xlabel("Numero accidentes")
plt.ylabel("Fechas")
plt.show()
```

```python
## Accidentes de trafico por mes
import re

def get_mes(fecha):
    return re.findall(r'\d+',fecha)[1]

accidentes = pd.Series([get_mes(date) for date in df.drop_duplicates(["num_expediente"], keep='first').fecha]).value_counts().to_dict()

months = ["ENERO", "FEBRERO", "MARZO", "ABRIL", "MAYO", "JUNIO", "JULIO", "AGOSTO", "SEPTIEMBRE", "OCTUBRE", "NOVIEMBRE", "DICIEMBRE"]
meses = [months[int(mes)] for mes in list(accidentes.keys())[0:10]]

plt.barh(meses, list(accidentes.values())[0:10], color=color_palette)
plt.title("Accidentes de trafico por meses")
plt.xlabel("Numero accidentes")
plt.ylabel("Mes")
plt.show()
```
