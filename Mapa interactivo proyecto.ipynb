```python
import folium

# Mapa de España
espana_map = folium.Map(location=[40.4637, -3.7492], zoom_start=6)  # Coordenadas centradas en España

# Marcadores para España y Comunidad de Madrid
folium.Marker([40.4168, -3.7038], popup='España: 1,145 muertos').add_to(espana_map)
folium.Marker([40.4168, -3.7038], popup='Madrid: 22 fallecidos').add_to(espana_map)

# Marcadores para otros países
folium.Marker([20.5937, 78.9629], popup='India: 356,000 muertos').add_to(espana_map)
folium.Marker([35.8617, 104.1954], popup='China: 248,000 muertos').add_to(espana_map)
folium.Marker([37.7749, -122.4194], popup='Estados Unidos: 38,800 muertos').add_to(espana_map)
folium.Marker([-14.2350, -51.9253], popup='Brasil: 35,000 muertos').add_to(espana_map)
folium.Marker([23.6345, -102.5528], popup='México: 30,000 muertos').add_to(espana_map)

# Mostrar mapa
espana_map
```
